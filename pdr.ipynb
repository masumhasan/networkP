```python
import traceanalyzer as tr
#Packet Delivery Ratio
pdr1=tr.Pdr('a.tr','33')
pdr1.sample()
pdr1.plot('sr-') #plotting with argument
#getting data
time=pdr1.time_sample
pdr=pdr1.pdr_sample
idx=0
for instant in time:
    print(instant,' ',pdr[idx]) 
    idx+=1 
```
